# t-test

# üß™ T-test
* Met√≥da matematickej ≈°tatistiky

## üß∞ Pou≈æitie
V praxi sa t-test ƒçasto pou≈æ√≠va na **porovnanie**, ƒçi sa **v√Ωsledky meran√≠** na **1. skupine v√Ωznamne l√≠≈°ia** od **v√Ωsledkov meran√≠ na 2. skupine**.

Umo≈æ≈àuje **overi≈•** niektor√∫ z nasleduj√∫cich **hypot√©z**:  
* H1.  ƒçi **1 norm√°lne rozdeleni**e, z ktor√©ho poch√°dza **urƒçit√Ω n√°hodn√Ω v√Ωber**, m√° **urƒçit√∫ konkr√©tnu stredn√∫ hodnotu/medi√°n**, priƒçom **rozptyl/variancia je nezn√°my/a**  
    * üï∫ v tomto pr√≠pade m√¥≈æe by≈• n√°hodn√Ω v√Ωber tvoren√Ω buƒè jednotliv√Ωmi hodnotami --> **jednov√Ωberov√Ω t-test** 
    * üë≠ alebo dvojicami hodn√¥t, pri ktor√Ωch sa sk√∫maj√∫ ich rozdiely --> **p√°rov√Ω t-test**      
* üëØ H2. ƒçi **2 norm√°lne rozdelenia** maj√∫ce **rovnak√Ω (hoci nezn√°my) rozptyl**, z ktor√Ωch poch√°dzaj√∫ 2 nez√°visl√© n√°hodn√© v√Ωbery, **maj√∫ rovnak√© stredn√© hodnoty/medi√°ny** resp. **rozdiel t√Ωchto stredn√Ωch hodn√¥t/medi√°nov** je **rovn√Ω urƒçit√©mu dan√©mu ƒç√≠slu** --> **dvojv√Ωberov√Ω t-test**

## ü§î Princ√≠p (Ako to funguje)
1. Ak **n√°hodn√Ω v√Ωber** poch√°dza z **norm√°lneho rozdelenia**, potom **v√Ωberov√Ω priemer/mean** m√° aj **norm√°lne rozdelenie** s **rovnakou strednou hodnotou/medi√°nom**
2. **Rozdiel v√Ωberov√©ho priemeru** a **strednej hodnoty** **normovan√Ω pomocou skutoƒçn√©ho rozptylu** by potom mal **norm√°lne rozdelenie** s **nulovou strednou hodnotou** a **jednotkov√Ωm rozptylom**. 
3. **Skutoƒçn√Ω rozptyl** v≈°ak **nie je zn√°my**. 
4. Ak je nahraden√Ω odhadom pomocou v√Ωberov√©ho rozptylu, vznikne **T rozdelenie**, ktor√© s **rast√∫cim poƒçtom stup≈àov voƒænosti** **konverguje (pribli≈æuje sa/zbieha) na norm√°lne rozdelenie**.

# üß™ 1-v√Ωberovy t-test (One Sample t-test)

## üë®‚Äçüè´ Interpret√°cia t-testu
1. **Testov√° ≈°tatistika t** teda **nadob√∫da hodnoty -3,1834** pri **9. stup≈àoch voƒænosti** 
2. Poƒçet stup≈àov voƒænosti df je v tomto teste o 1 ni≈æ≈°iu ako poƒçet meran√≠ N (df = N - 1)
3. Zodpovedaj√∫ca **p-hodnota je 0,01112**, ƒço je **men≈°ie ako 0,05** a tak m√¥≈æeme **0 hypot√©zu na zadanej hladine v√Ωznamnosti zamietnu≈•**
4. M√¥≈æeme poveda≈•, ≈æe **stredn√° hodnota/medi√°n v√Ω≈°ky mu≈æov kme≈àa Apaƒçov nie je 175 cm**, ale zrejme o nieƒço men≈°ia
5. Program vyp√≠sal **aritmetick√Ω priemer v√Ω≈°ok meranej skupiny 165,7 cm** a jeho 95 % konfidenƒçn√Ω interval pribli≈æne 159,1 a≈æ 172,3 cm
