# Parsovanie XML d√°t
- XML (eXtensible Markup Language) je znaƒçkovac√≠ jazyk pou≈æ√≠van√Ω na reprezent√°ciu hierarchick√Ωch d√°t. 
- Na parsovanie XML v Pythone m√¥≈æeme pou≈æi≈• kni≈ænicu xml.etree.ElementTree.

```python
import xml.etree.ElementTree as ET

xml_data = '''<?xml version="1.0"?>
<data>
    <osoba>
        <meno>Adam</meno>
        <priezvisko>≈†angala</priezvisko>
        <vek>33</vek>
        <mesto>Trnava</mesto>
    </osoba>
</data>'''

# root
koren = ET.fromstring(xml_data)

for osoba in koren.findall('osoba'):
    meno = osoba.find('meno').text
    vek = osoba.find('vek').text
    mesto = osoba.find('mesto').text
    print(f'Meno: {meno}, Vek: {vek}, Mesto: {mesto}')
```

# Naƒç√≠tanie XML s√∫boru a vyp√≠sanie jeho obsahu

```python
import xml.etree.ElementTree as ET

# Naƒç√≠tanie XML s√∫boru
tree = ET.parse('krajiny.xml')
root = tree.getroot()

# Vyp√≠sanie obsahu
for country in root.findall('country'):
    country_name = country.get('name')
    print(f'Krajina: {country_name}')
    for city in country.findall('city'):
        city_name = city.get('name')
        population = city.get('population')
        print(f' Mesto: {city_name}, Popul√°cia: {population}')
```

# Pridanie nov√©ho elementu do existuj√∫ceho XML dokumentu

# Parsovanie URL Adries
* Pri pr√°ci s webov√Ωmi aplik√°ciami ƒçasto potrebujeme analyzova≈• alebo kon≈°truova≈• URL adresy. 
* Python poskytuje modul urllib na pr√°cu s URL.

```python
from urllib.parse import urlparse, parse_qs

url = 'https://www.priklad.sk/index.php?id=123&category=hlavna'

parsovana_url = urlparse(url)
dopyt_parametre = parse_qs(parsovana_url.query)

print(f'Path: {parsovana_url.path},\nDopyt (Query) Parametre: {dopyt_parametre}')

url2 = 'https://www.vita.sk/?s=python'

parsovana_url2 = urlparse(url2)
dopyt_parametre2 = parse_qs(parsovana_url2.query)

print("----------------------\n")
print(f'Path: {parsovana_url2.path},\nDopyt (Query) Parametre: {dopyt_parametre2}')

url3 = 'https://www.vita.sk/online-kurzy-programovacie-jazyky/?orderby=&filter%5Bpa_uroven-obtiaznosti-kurzu%5D%5B%5D=253&filter%5Bpa_tema%5D%5B%5D=279'

parsovana_url3 = urlparse(url3)
dopyt_parametre3 = parse_qs(parsovana_url3.query)

print("----------------------\n")
print(f'Path: {parsovana_url3.path},\nDopyt (Query) Parametre: {dopyt_parametre3}')
```

# Parsovanie Log S√∫borov
* Log s√∫bory obsahuj√∫ veƒæk√© mno≈æstvo d√°t generovan√Ωch aplik√°ciami, ktor√© m√¥≈æu by≈• u≈æitoƒçn√© pre anal√Ωzu a monitorovanie. 
* Tieto s√∫bory ƒçasto maj√∫ ≈°pecifick√Ω form√°t, ktor√Ω je mo≈æn√© analyzova≈• regul√°rnymi v√Ωrazmi v Pythone.

```python
import re

log_data = """
INFO 2024-04-01 12:00:01 Vzorova Sluzba - Start procesu
ERROR 2024-04-01 12:01:35 Vzorova Sluzba - Chyba pri spracovan√≠ √∫dajov
ERROR 2024-04-01 12:01:48 Vzorova Sluzba - Chyba pri spracovan√≠ √∫dajov
INFO 2024-04-01 12:02:12 Vzorova Sluzba - Proces dokonƒçen√Ω √∫spe≈°ne
"""

pattern = re.compile(r'ERROR (.*) - (.*)')
for riadok in log_data.split('\n'):
    zhoda = pattern.search(riadok)
    if zhoda:
        print(f'ƒåas: {zhoda.group(1)}, Spr√°va: {zhoda.group(2)}')
```

# Parsovanie komplexn√Ωch textov√Ωch d√°t pomocou regul√°rnych v√Ωrazov
Regul√°rne v√Ωrazy s√∫ mocn√Ωm n√°strojom na parsovanie a manipul√°ciu s textov√Ωmi re≈•azcami, umo≈æ≈àuj√∫ci identifikova≈• vzory v texte.

```python
import re

text = "Email: miroslav.reiter@gmail.com, Phone: +421908123456"
email_pattern = r"[\w\.-]+@[\w\.-]+"
telefon_pattern = r"\+\d{12}"

email = re.search(email_pattern, text).group()
telefon = re.search(telefon_pattern, text).group()

print(f'Email: {email}, Telef√≥n: {telefon}')
```

# Parsovanie Titulky (Title)

```python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>Moja str√°nka</title></head>
<body><p class="nadpis">Vitajte na mojej str√°nke!</p></body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.title.text)  # Vyp√≠≈°e: Moja str√°nka
```

# Parsovanie Nadpisy (Headings)

```python
from bs4 import BeautifulSoup

# Simulovan√© HTML dokumentu
html_doc = """
<html>
<head>
<title>N√°zov str√°nky</title>
</head>
<body>
<h1>M√¥j prv√Ω nadpis</h1>
<p>M√¥j prv√Ω odstavec.</p>
<h2>Druh√Ω nadpis</h2>
<p>Druh√Ω odstavec.</p>
</body>
</html>
"""

# Parsovanie HTML
soup = BeautifulSoup(html_doc, 'html.parser')

# Vyhƒæad√°vanie v≈°etk√Ωch nadpisov h1 a h2
nadpisy = soup.find_all(['h1', 'h2'])

for nadpis in nadpisy:
    print(nadpis.text)
```

# Parsovanie Odkazov (Anchors)

```python
from bs4 import BeautifulSoup

# Simulovan√© HTML s odkazmi
html_doc = """
<html>
<body>
<a href="https://google.sk/prvy" class="sibling">1. Odkaz</a>
<a href="https://google.sk/druhy" class="sibling">2. Odkaz</a>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Vyhƒæad√°vanie v≈°etk√Ωch odkazov
odkazy = soup.find_all('a')

for odkaz in odkazy:
    print(odkaz['href'])  # Vyp√≠≈°e URL odkazu
```

# Parsovanie Produktov (class, DIV a P)

```python
from bs4 import BeautifulSoup

# Simulovan√© HTML
html_doc = """
<div class="product">
    <p class="name">Produkt 1</p>
    <p class="price">Cena: 100‚Ç¨</p>
</div>
<div class="product">
    <p class="name">Produkt 2</p>
    <p class="price">Cena: 200‚Ç¨</p>
</div>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Vyhƒæad√°vanie pomocou CSS selektorov
products = soup.select('.product .name')

for product in products:
    print(product.text)  # Vyp√≠≈°e n√°zvy produktov
```

# Zmena Obsahu a Sprehƒæadnenie (Prettify)

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
<body>
<p id="uvod">Toto je uvodn√Ω odstavec.</p>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Zmena textu v elemente
uvod = soup.find('p', id='uvod')
uvod.string = "Toto je zmenen√Ω uvodn√Ω odstavec."

print(soup.prettify())
```

# Vymazanie Obsahu

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
<body>
<div>
<p>Toto je odstavec, ktor√Ω zostane.</p>
<p id="zmazat">Toto je odstavec, ktor√Ω bude odstr√°nen√Ω.</p>
</div>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Odstr√°nenie elementu
zmazat = soup.find('p', id='zmazat')
zmazat.decompose()

print(soup.prettify())
```

# Vyhƒæad√°vanie Viacer√Ωch Elementov (UL OL a LI)

```python
from bs4 import BeautifulSoup

html_doc = """
<ul>
<li class="polozka">Jablko</li>
<li class="polozka">Ban√°n</li>
<li class="polozka">Pomaranƒç</li>
</ul>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Vyhƒæad√°vanie viacer√Ωch elementov
polozky = soup.find_all('li', class_='polozka')

for polozka in polozky:
    print(polozka.text)  # Vyp√≠≈°e text ka≈ædej polo≈æky
```

# Pridanie Nov√©ho Elementu

```python
from bs4 import BeautifulSoup

html_doc = """
<div id="kontajner">
<p>Prv√Ω odstavec.</p>
</div>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

# Vytvorenie nov√©ho elementu
novy_odstavec = soup.new_tag('p')
novy_odstavec.string = "Toto je nov√Ω odstavec."

# Pridanie nov√©ho elementu do kontajnera
kontajner = soup.find('div', id='kontajner')
kontajner.append(novy_odstavec)

print(soup.prettify())
```

# Naƒç√≠tanie Cel√©ho Obsahu Webovej Str√°nky

```python
import requests
from bs4 import BeautifulSoup

# Adresa webovej str√°nky, ktor√∫ chceme spracova≈•
url = 'https://www.vita.sk'

# Po≈°leme GET po≈æiadavku na server
response = requests.get(url)

# Vytvor√≠me objekt BeautifulSoup pre anal√Ωzu z√≠skan√©ho HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Vytlaƒç√≠me cel√© HTML na kontrolu
print(soup.prettify())
```

# N√°jdenie Podnadpisov

```python
# N√°jdeme v≈°etky elementy 'h2'
nadpisy = soup.find_all('h2')

for nadpis in nadpisy:
    print(nadpis.text)
```

# N√°jdenie V≈°etk√Ωch Obr√°zkov

```python
# N√°jdenie v≈°etk√Ωch obr√°zkov a vytlaƒçenie ich zdrojov
obrazky = soup.find_all('img')

for obrazok in obrazky:
    src = obrazok.get('src')
    print(src)
```

# N√°jdenie V≈°etk√Ωch Paragrafov obsahuj√∫cich slovo Kurz

```python
# N√°jdenie paragrafov obsahuj√∫cich slovo 'kurz'
paragrafy = soup.find_all('p')

for p in paragrafy:
    if 'kurz' in p.text:
        print(p.text)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generujeme n√°hodn√© d√°ta
X = np.random.rand(100, 2)

# Pou≈æijeme K-means algoritmus na klastrovanie d√°t do 3 skup√≠n
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Vizualizujeme klastre a centr√° klastrov
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.title("Vizualiz√°cia v√Ωsledkov klastrovania")
plt.xlabel("X hodnota")
plt.ylabel("Y hodnota")
plt.show()
```

# Pyparsing 

# Pyparsing
* umo≈æ≈àuje jednoduch√∫ a intuit√≠vnu anal√Ωzu a manipul√°ciu s textov√Ωmi d√°tami bez nutnosti vytv√°ra≈• komplikovan√© gramatiky alebo sa zapodieva≈• do detailov regul√°rnych v√Ωrazov. 
* Je obzvl√°≈°≈• u≈æitoƒçn√° pre anal√Ωzu d√°t a pre transform√°ciu d√°tov√Ωch form√°tov. 

```python
from pyparsing import Word, nums, oneOf

# Defin√≠cia z√°kladn√Ωch stavebn√Ωch blokov
number = Word(nums)  # ƒå√≠sla
operator = oneOf("+ - * /")  # Oper√°tory

# Vytvorenie gramatiky
expression = number + operator + number

# Analyzovanie v√Ωrazu
result = expression.parseString("3 + 4")

print(result)  # V√Ωstup: ['3', '+', '4']
```

# Anal√Ωza Logick√©ho V√Ωrazu
Ak by sme chceli analyzova≈• trochu zlo≈æitej≈°√≠ logick√Ω v√Ωraz obsahuj√∫ci z√°tvorky a logick√© oper√°tory

```python
from pyparsing import Word, alphas, opAssoc, infixNotation

# Defin√≠cia z√°kladn√Ωch prvkov
variable = Word(alphas)
and_ = "a"
or_ = "alebo"
not_ = "nie"

# Defin√≠cia gramatiky pre logick√Ω v√Ωraz
expr = infixNotation(variable,
    [
        (not_, 1, opAssoc.RIGHT),
        (and_, 2, opAssoc.LEFT),
        (or_, 2, opAssoc.LEFT),
    ])

# Analyzovanie v√Ωrazu
result = expr.parseString("A a nie B alebo C")

print(result.asList())  # V√Ωstup uk√°≈æe vnoren√∫ ≈°trukt√∫ru v√Ωrazu
print(result)  # V√Ωstup 
print(expr)  # V√Ωstup 
```

# Anal√Ωza a Transform√°cia D√°t
Pyparsing je tie≈æ skvel√Ω na transform√°ciu textov√Ωch d√°t. Predstavme si, ≈æe chceme transformova≈• d√°tov√Ω z√°znam zo zlo≈æit√©ho form√°tu na slovn√≠k

```python
from pyparsing import Word, alphas, alphanums, Suppress, Group, Dict

# Elementy gramatiky
identifier = Word(alphas, alphanums + "_").setName("identifier")
equals = Suppress("=")
value = Word(alphanums).setName("value")

# Pravidlo pre z√°znam
record = Group(identifier + equals + value)

# Defin√≠cia cel√©ho vstupu
config = Dict(record[...])

# Anal√Ωza a transform√°cia d√°t
data = "meno=Adam vek=33"
result = config.parseString(data)

print(dict(result))  # V√Ωstup: {'meno': 'Adam', 'vek': '33'}
```

```python
from pyparsing import Word, alphanums, Combine, Literal

# Defin√≠cia ƒçast√≠ emailovej adresy
localPart = Word(alphanums + "._")
atSymbol = Literal("@")
domain = Word(alphanums + ".")
emailAddress = Combine(localPart + atSymbol + domain)

# Anal√Ωza a transform√°cia
email = "miroslav.reiter@gmail.com"
transformed = emailAddress.transformString("Osobn√Ω email: miroslav.reiter@gmail.com")
transformed2 = emailAddress.transformString("Zastihnete ma na firemnom emaily riaditel@itacademy.sk alebo v Bratislave")

print(transformed)  # V√Ωstup: 'Contact: example.user@domain.com'
print(transformed2)  # V√Ωstup: 'Contact: example.user@domain.com'
```

```python
from pyparsing import oneOf, OneOrMore, Word, alphas

# Preddefinovan√© kl√∫ƒçov√© slov√°
keywords = oneOf("Python Pyparsing anal√Ωza d√°ta transform√°cia")

# Pravidlo pre slov√°, ktor√© nie s√∫ kl√∫ƒçov√© slov√°
nonKeyword = Word(alphas)

# Gramatika, ktor√° ignoruje v≈°etko okrem kl√∫ƒçov√Ωch slov
grammar = OneOrMore(keywords | nonKeyword.suppress())

# Analyzovanie textu
text = "Pyparsing je kni≈ænica v Python pre anal√Ωza a transform√°cia d√°ta."
result = grammar.searchString(text)

print(result)  # V√Ωstup zobraz√≠ len kl√∫ƒçov√© slov√°
```

```python
from pyparsing import makeHTMLTags, SkipTo

# Vytvorenie pravidiel pre HTML link tag
anchorStart, anchorEnd = makeHTMLTags("a")
link = anchorStart + SkipTo(anchorEnd).setResultsName("url") + anchorEnd

# Pr√≠klad HTML textu
htmlText = '<a href="https://google.sk">Stranka Google.sk</a>'

# Vyhƒæad√°vanie a extrakcia linku
for token, start, end in link.scanString(htmlText):
    print(token.url)  # V√Ωstup: 'Stranka Google.sk'
```
